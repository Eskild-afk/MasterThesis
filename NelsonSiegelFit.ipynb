```python
from scipy.optimize import minimize, basinhopping, fsolve
from NelsonSiegel import NelsonSiegel



def NSoptimizer(X1,X2,X3,Lambda, sigmaL, sigmaS, sigmaC, observedP, observedK):
    if (np.array([sigmaL, sigmaS, sigmaC])<0).any():
        print('Volatility is negative', end='r')
        return 100000
    
    if X1+X2+X3<0.02:
        return 100000
    
    if X1+X2+X3>0.03:
        return 100000
    straddleModel = observedP.copy()
    optNS = NelsonSiegel(initial=np.array([X1,X2,X3]),
                         Lambda=Lambda,
                         sigmaL=sigmaL,
                         sigmaS=sigmaS,
                         sigmaC=sigmaC
                        )

    for expiry in observedP.index:
        # Expiry first
        # Getting the right expiry and converting to years
        # Simulate the rates here:
        np.random.seed(1337)
        N = 10000
        rate = [optNS.oneStep(0,X1+X2+X3, expiry, expiry) for _ in range(N)]


        for tenor in observedP.columns:
            # Creating tenor legs
            T = np.arange(expiry/12, expiry/12+tenor/12+0.5, 0.5)
            S = np.arange(expiry/12, expiry/12+tenor/12+1)
            # Finding the ATM fixed rate
            K = observedK[tenor].loc[expiry]/100
            # K = fsolve(lambda x: optHW.swap(0, S, T, x), x0=MarketATM)[0]
            swap = np.array([float(optNS.swap(expiry, S, T, K, x, payer=True)) for x in rate])
            #Calculating the payer and receiver swpt price
            p = optNS.ZCB(0,expiry)*np.maximum(swap,0).mean()
            r = -optNS.ZCB(0,expiry)*np.minimum(swap,0).mean()
            
            #Normalizing w.r.t. to the A(t,S) function (2.10) everythin you wanna know...
            A = np.array([optNS.ZCB(0, Si) for Si in S[1::]]).sum()

            straddleModel[tenor].loc[expiry] = (p+r)/A
            # if (p+r)*10000<0:
            # print(p,r,expiry,tenor,reversion,volatility)
            

    error = ((straddleModel/observedP-1)**2).sum().sum()
    print(f'{error}', end='                                                         \r')
    return error
```
